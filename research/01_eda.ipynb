# 01 — Exploratory Data Analysis
Understand the raw Online Retail II dataset before any cleaning.

## 1. Imports & config

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:.2f}".format)
sns.set_theme(style="whitegrid", palette="muted")
%matplotlib inline
```

## 2. Load raw data

```python
RAW = Path("../data/raw/Online_Retail.csv") # .. means go up one directory level, then into data/raw

df = pd.read_csv(RAW,encoding='latin1')

print(df.shape)
df.head()
```

## 3. Basic info

```python
df.info()
```

```python
df.describe(include="all")
```

## 4. Missing values

```python
missing = df.isnull().sum().rename("Missing").to_frame()
missing["Pct"] = (missing["Missing"] / len(df) * 100).round(2)
missing[missing["Missing"] > 0].sort_values("Pct", ascending=False)
```

## 5. Duplicate rows

```python
dupes = df.duplicated().sum()
print(f"Duplicate rows: {dupes:,} ({dupes/len(df)*100:.2f}%)")
```

## 6. Cancellations

```python
df["IsCancelled"] = df["InvoiceNo"].astype(str).str.startswith("C")
print(df["IsCancelled"].value_counts())
print(f"Cancellation rate: {df['IsCancelled'].mean()*100:.2f}%")
```

## 7. Date range

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
print("Earliest:", df["InvoiceDate"].min())
print("Latest  :", df["InvoiceDate"].max())
print("Span    :", (df["InvoiceDate"].max() - df["InvoiceDate"].min()).days, "days")
```

## 8. Revenue distribution

```python
df_valid = df[~df["IsCancelled"] & df["CustomerID"].notna()].copy()
df_valid["Revenue"] = df_valid["Quantity"] * df_valid["UnitPrice"]

fig, axes = plt.subplots(1, 2, figsize=(14, 4))
df_valid["Revenue"].clip(0, 500).hist(bins=60, ax=axes[0], color="steelblue", edgecolor="white")
axes[0].set_title("Revenue per line item (clipped at £500)")
axes[0].set_xlabel("Revenue (£)")

df_valid.groupby("InvoiceNo")["Revenue"].sum().clip(0, 2000).hist(bins=60, ax=axes[1], color="coral", edgecolor="white")
axes[1].set_title("Order value distribution (clipped at £2000)")
axes[1].set_xlabel("Order Total (£)")
plt.tight_layout()
plt.show()
```

## 9. Top 10 countries by revenue

```python
top_countries = (
    df_valid.groupby("Country")["Revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
top_countries.plot(kind="barh", figsize=(10, 5), color="steelblue", edgecolor="white")
plt.title("Top 10 Countries by Revenue")
plt.xlabel("Total Revenue (£)")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## 10. Monthly revenue trend

```python
df_valid["YearMonth"] = df_valid["InvoiceDate"].dt.to_period("M")
monthly = df_valid.groupby("YearMonth")["Revenue"].sum()

monthly.plot(figsize=(12, 4), marker="o", color="steelblue")
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue (£)")
plt.tight_layout()
plt.show()
```

## 11. Orders by day of week

```python
df_valid["DayOfWeek"] = df_valid["InvoiceDate"].dt.day_name()
order_day = df_valid.groupby("DayOfWeek")["InvoiceNo"].nunique()
order_day = order_day.reindex(["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"])

order_day.plot(kind="bar", figsize=(10, 4), color="mediumpurple", edgecolor="white")
plt.title("Unique Orders by Day of Week")
plt.ylabel("Order Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 12. Customer coverage

```python
total_customers = df["CustomerID"].nunique()
customers_with_id = df_valid["CustomerID"].nunique()
print(f"Total unique CustomerIDs (incl nulls): {total_customers:,}")
print(f"Customers with valid ID             : {customers_with_id:,}")
print(f"Coverage                            : {customers_with_id/total_customers*100:.1f}%")
```

## Key Observations

- The dataset contains **541,909 transactions**, spanning **373 days** from **Dec 2010 to Dec 2011**, across **38 countries**.
- **CustomerID** has **24.93% missing values**, making those records unsuitable for customer-level analysis, while missing descriptions are negligible (**0.27%**).
- Approximately **0.97% of rows are duplicates** and should be removed during data cleaning.
- **1.71% of transactions are cancellations** (InvoiceNo starting with 'C') and must be excluded from revenue and customer behavior analysis.
- The dataset contains **4,339 valid customers**, providing sufficient data for segmentation and retention studies.
- Quantity and UnitPrice contain **extreme positive and negative values**, indicating returns, cancellations, and potential outliers that require treatment.
- Revenue and order value distributions are **highly right-skewed**, with a small number of customers/orders contributing disproportionately high revenue.
- Most transactions are low-value purchases, while a small segment of customers exhibits significantly higher spending behavior.
- Overall, the dataset is well-suited for **RFM analysis, customer segmentation, CLV estimation, and retention modeling** after appropriate cleaning and preprocessing.
